```python
f1=input ("Digite uma Frase: ")
###############################
Fc=len(f1)
QP=f1.split()
quantp=len(QP)
mc=float(Fc/quantp)
mcr=round(mc,2)
###############################
vogais="aeiouAEIOU"
qvogais=0
for letras in f1:
  if letras in vogais:
    qvogais+=1
###############################
pv=float((qvogais/Fc)*100)
pvr=round(pv,2)
###############################
print("sua frase tem ",Fc,"caracteres")
print("percentual de vogais",pvr,"%")
print("sua frase tem ",quantp,"palavras")
print("em média temos",mcr,"caracteres por palavra")
```
